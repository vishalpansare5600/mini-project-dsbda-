```python
negative=df[df['sentiment']=='negative']
wc=WordCloud(max_font_size=50, max_words=100, background_color="white").generate(str(negative['text']))
plt.figure(figsize=(10,10))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
